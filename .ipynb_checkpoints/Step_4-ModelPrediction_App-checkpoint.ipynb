### In the next step we will graph all features and separate them between Benign and Malignant. We want to see the final features that have the highest correlation to Dx. 

```python
# Step 1: Separate out M and B data for graph comparison
malignant = df[df['diagnosis'] ==1]
benign = df[df['diagnosis'] ==0]
```

```python
#Select column names based on our correlation analysis above of which features were highly correlated with Dx
list = ['radius_mean']+ ['perimeter_mean']+['area_mean']+['concavity_mean']+['concave_points_mean'] + ['radius_worst'] + ['perimeter_worst'] +['area_worst']+ ['concavity_worst']+ ['concave_points_worst']
correlation_data = df.loc[:,list]
correlation_data.head()
```

```python
correlation_data['diagnosis'] = df['diagnosis']
correlation_data.head()
```

```python
correlation_data.columns
```

```python
df_feature_only = df.drop("diagnosis", axis=1)
df_list=df_feature_only.columns
```

```python
#Plot all 30 features to show correlations 
plt.rcParams.update({'font.size': 10})
plot, graphs = plt.subplots(nrows=15, ncols=2, figsize=(20,24))
graphs = graphs.flatten()
for idx, graph in enumerate(graphs):
    graph.figure
    
    binwidth= (max(df[df_list[idx]]) - min(df[df_list[idx]]))/50
    bins = np.arange(min(df[df_list[idx]]), max(df[df_list[idx]]) + binwidth, binwidth)
    graph.hist([malignant[df_list[idx]],benign[df_list[idx]]], bins=bins, alpha=0.6, density=True, label=['Malignant','Benign'], color=['red','green'])
    graph.legend(loc='upper right')
    graph.set_title(df_list[idx])
plt.tight_layout()
plt.subplots_adjust(top=0.88)
```

```python
#Plots of 10 features with the highest correlations based on above analysis
plt.rcParams.update({'font.size': 10})
plot, graphs = plt.subplots(nrows=5, ncols=2, figsize=(14,16))
graphs = graphs.flatten()

for idx, graph in enumerate(graphs):
    graph.figure
    
    binwidth= (max(df[list[idx]]) - min(df[list[idx]]))/50
    bins = np.arange(min(df[list[idx]]), max(df[list[idx]]) + binwidth, binwidth)
    graph.hist([malignant[list[idx]],benign[list[idx]]], bins=bins, alpha=0.6, density=True, label=['Malignant','Benign'], color=['red','green'])
    graph.legend(loc='upper right')
    graph.set_title(list[idx])
plt.tight_layout()
plt.subplots_adjust(top=0.88)
```

### We will use Principle Component Analysis (PCA) for feature extraction. 
Prior to PCA, we need to normalize data for better performance of PCA


```python
# Assign X (data) and y (target)

### BEGIN SOLUTION
X_data = df.drop("diagnosis", axis=1)
y_data = df["diagnosis"]
print(X_data.shape, y_data.shape)
### END SOLUTION
```

```python
# Split data train 70 % and test 30 %
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.3, random_state=42)
```

```python
#Normalization process
X_train_N = (X_train-X_train.mean())/(X_train.max()-X_train.min())
X_test_N = (X_test-X_test.mean())/(X_test.max()-X_test.min())
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_train_N)

plt.figure(1, figsize=(10, 8))
plt.clf()
plt.axes([.2, .2, .7, .7])
plt.plot(pca.explained_variance_ratio_, linewidth=2)
plt.axis('tight')
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_')
```

The variance ration indicates that 5 component can be chosen


# Model Classification
Based on our dataset, we will try 3 classification models to determine which will be most successful at predicting future outcomes based on user input. 
- Trees(RandomForestClassifier) with 10 selected features¶


# Trees(RandomForestClassifier)
### 10 selected features

```python
target =correlation_data["diagnosis"]
target_names = ["Benign", "Malignant"]
```

```python
data = correlation_data.drop("diagnosis", axis=1)

feature_names = data.columns
data.head()
```

```python
#Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
# Fit to the training data and validate with the test data
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
# # Use this to visualize the tree
# import graphviz 
# dot_data = tree.export_graphviz(
#     clf, out_file=None, 
#     feature_names=feature_names,  
#     class_names=target_names,  
#     filled=True, rounded=True,  
#     special_characters=True)  

# graph = graphviz.Source(dot_data)  
# graph
```

```python
# Fit to the training data and validate with the test data
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

#### By looking at the confusion matrix below we can see how accurate the obtained results are

```python
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test,rf.predict(X_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test, rf.predict(X_test))
sns_plot=sns.heatmap(cm,annot=True,fmt="d", cmap='coolwarm')
sns_plot.figure.savefig("RandomForest.png")
```

# KNN
### 10 selected features

```python
# Loop through different k values to see which has the highest accuracy
# Only use odd numbers to avoid ties
train_scores = []
test_scores = []
for k in range(1, 20, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 20, 2), train_scores, marker='o')
plt.plot(range(1, 20, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
#plt.savefig('KNN.png')
plt.show()
```

```python
# K: 11 seems to be the most appropriate choice for this dataset
knn = KNeighborsClassifier(n_neighbors=11)
knn.fit(X_train, y_train)
print('k=11 Test Acc: %.3f' % knn.score(X_test, y_test))
```

# SVM (Support Vector Machine)
### 10 selected features¶

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Model Accuracy
print('Test Acc: %.3f' % model.score(X_test, y_test))
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
print(classification_report(y_test,  y_pred,
                            target_names=target_names))
print('Accuracy is: ', accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns_plot=sns.heatmap(cm,annot=True,fmt="d", cmap = 'coolwarm')
#sns_plot.figure.savefig("SVM.png")
```

### Based on the models above, the best model for prediction (detection of breast cancer types) is SVM which also has a great F1 score 0.99
